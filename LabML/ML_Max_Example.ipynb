Neural Network (MLPRegressor)

```python
import random as rnd
import numpy as np

n = 10
Nexamples = 1000

X = []
y = []
for j in range(Nexamples):
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))
    y.append(max(list1))  # ModificaciÃ³n: almacenar el valor mÃ¡ximo en lugar de una lista ordenada

import pandas as pd
data = pd.DataFrame({'X': [a for a in X], 'y': [a for a in y]})
data

```

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos de entrenamiento
X_train = np.array(X)
y_train = np.array(y)

# Entrenamiento del modelo MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(100, 50, 20), random_state=1, max_iter=1000, activation='relu').fit(X_train, y_train)
y_pred = model.predict(X_train)

# Graficar la curva
plt.plot(model.loss_curve_)
plt.show()

# Calcular el error (diferencia absoluta entre predicciones y valores reales)
error = np.abs(y_train - y_pred)

# Crear el DataFrame con las predicciones y los errores
y_pred_df = pd.DataFrame({
    'X_train': [a for a in X_train],
    'y_true': y_train,
    'y_pred': [round(b, 3) for b in y_pred],
    'error': [round(e, 3) for e in error]
})

y_pred_df.head()
```

```python
import random as rnd
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import pandas as pd

n = 10
Nexamples = 1000

X = []
y = []

# Generar datos para valores de n normales
for j in range(Nexamples):
    y.append(max(np.array(range(n))))
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))

# Generar datos para valores de n multiplicados por 2
for j in range(Nexamples):
    y.append(max(2 * np.array(range(n))))
    list1 = list(2 * np.array(range(n)))
    rnd.shuffle(list1)
    X.append(np.array(list1))

model = MLPRegressor(hidden_layer_sizes=(100, 50, 20), max_iter=1000, activation='tanh').fit(X, y)
y_pred = model.predict(X)

plt.plot(model.loss_curve_)
plt.show()

print('Accuracy testing : {:.3f}'.format(model.score(X, y)))

# Calcular el error entre las predicciones y los valores reales
error = np.abs(y - y_pred).mean()

data = pd.DataFrame({'X': [a for a in X], 'y': [a for a in y], 'y_pred': [np.round(a, 3) for a in y_pred], 'error': round(error, 3)})
data


```

```python
import random as rnd
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import pandas as pd

n = 10
Nexamples = 10000

X = []
y = []

for j in range(Nexamples):
    y.append(np.array(range(n)))
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))

for j in range(Nexamples):
    y.append(2 * np.array(range(n)))
    list1 = list(2 * np.array(range(n)))
    rnd.shuffle(list1)
    X.append(np.array(list1))

model = MLPRegressor(hidden_layer_sizes=(100, 50, 20), max_iter=1000, activation='tanh').fit(X, y)
y_pred = model.predict(X)

# Obtener el valor mÃ¡ximo en y_pred
y_pred_max = np.max(y_pred, axis=1)

# Calcular el error absoluto entre y_pred_max y el valor mÃ¡ximo real en y
error = np.abs([np.max(y[i]) for i in range(len(y))] - y_pred_max)

print('Accuracy testing : {:.3f}'.format(model.score(X, y)))

data = pd.DataFrame({'X': [a for a in X], 'y': [a for a in y], 'y_pred_max': y_pred_max, 'error': error})
data
```

Classification and Regression Tree (CART)

```python
import random as rnd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

n = 10
Nexamples = 10000

X = []
y = []

for j in range(Nexamples):
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))
    y.append(np.max(np.array([rnd.randint(0, 10) for a in range(n)])))

for j in range(Nexamples):
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))
    y.append(np.max(list1))

for j in range(Nexamples):
    list1 = list(2 * np.array(range(n)))
    rnd.shuffle(list1)
    X.append(np.array(list1))
    y.append(np.max(list1))

model = DecisionTreeRegressor(max_leaf_nodes=15).fit(X, y)
y_pred = model.predict(X)

data = pd.DataFrame({
    'X': [a for a in X],
    'y': [a for a in y],
    'y_pred': [a for a in y_pred],
    'error': [abs(a - b) for a, b in zip(y, y_pred)]
})
data
```
